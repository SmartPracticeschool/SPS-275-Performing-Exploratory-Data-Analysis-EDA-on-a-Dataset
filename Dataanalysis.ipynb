```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv('hour.csv')
dataset.head()
```

```python
dataset.shape
```

```python
dataset.index=dataset['dteday']
dataset=dataset.drop(columns=['dteday'])
dataset.head()
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
dataset.isnull().sum()
```

```python
dataset.plot(subplots=True)
```

```python
col=['season','yr','mnth','hr','weekday','workingday','weathersit']
for  i in col:
    print(i,'  has ',dataset[i].unique(),'  unique values')
```

```python
sns.heatmap(dataset.isnull())
```

```python
temp=dataset['temp']
print(type(temp))
```

```python
temp=np.array(temp)
print(temp.ndim)
print(temp.shape)
print(temp.dtype)
```

```python
temp=np.reshape(temp,(-1,1))
print(temp.shape)
print(temp.ndim)
```

```python
print('mean of temp is  :',np.mean(temp))
print('median of temp is  :',np.median(temp))
print('standard deviation of temp is ',np.std(temp))
```

```python
print('minimum value of temp is  :',temp.min())
print('maximum value of temp is  :',temp.max())
print('maximum value at position  :',temp.argmax())
print('minimum value at position  :',temp.argmin())
```

```python
hum=dataset['hum']
hum=np.array(hum)
hum=np.reshape(hum,(-1,1))
print(hum.shape)
print(hum.ndim)
```

```python
hum=hum.flatten()
temp=hum.flatten()
dot_result=np.dot(temp,hum)
print(dot_result)
```
